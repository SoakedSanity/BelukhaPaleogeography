```python
#Imports libraries needed for the project

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Forms and unpacks a utility "raw.txt" file from 18O isotope and temperature anomaly reconstruction dataset.

with open("raw.txt", "w", encoding = 'ansi') as f:
    f.write('')
f.close()

with open("belukha2009.txt", "r", encoding = 'ansi') as f:
    rows = f.readlines()
f.close()

with open("raw.txt", "a", encoding = 'ansi') as f:
    for k in range(100, len(rows)):
       f.write('   '.join(rows[k].split()) + "\n")
        
f.close()


years1, o18, tanom, sigma  = np.loadtxt('raw.txt', unpack=True)
```

```python
#Create a function for the graph smoothening using a moving average method.

def runmean(years, d):
    series = 5
    d_runmean = np.convolve(d, np.ones(series)/series, mode='valid')
    return (years[(series-1)//2:-(series-1)//2], d_runmean)
```

```python
#Forms and unpacks a utility "raw.txt" file from the major ion concentrations dataset.

with open('raw.txt', 'w', encoding = 'ansi') as p:
    p.write('')
p.close()

with open('belukha2009ions.txt', 'r', encoding = 'ansi') as o:
    rows_1 = o.readlines()
o.close

with open('raw.txt', 'a', encoding = 'ansi') as d:
    for i in range(132, len(rows_1)):
        d.write('   '.join(rows_1[i].split()) + "\n")
d.close()

years2, formate, chloride, nitrate, sulfate, sodium, ammonium, pottassium, magnesium, calcium = np.loadtxt('raw.txt', unpack=True)
```

```python
#Creates masks for making different dataset x-axis dimensions equal (both start at 1255 but end differently).

mask1 = years1 <= min(max(years1), max(years2)) 
mask2 = years2 <= min(max(years1), max(years2)) 
```
